```python
import time

import requests

from topix.utils.common import gen_uid
```

```python
endpoint = "http://localhost:8688"
```

```python
# Create a chat
st = time.time()
res = requests.put(f"{endpoint}/chats", params={"user_id": "root"})
print(f"Time lapsed for creating chat: {time.time() - st:.2f} seconds")
print(res.json())
```

```python
chat_id = res.json()["data"]["chat_id"]
```

```python
res = requests.post(
    f"{endpoint}/chats/{chat_id}/messages",
    params={"user_id": "root"},
    json={
        "query": "How to create a class with python?",
        "message_id": gen_uid(),
        "activated_tool": "code_interpreter",
    },
    stream=True
)

for chunk in res.iter_lines():
    if chunk:
        print(chunk.decode("utf-8"))
```

```python
res = requests.post(f"{endpoint}/chats/{chat_id}:describe", params={"user_id": "root"})
```
